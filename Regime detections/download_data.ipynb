```python
import yfinance as yf
import pandas as pd

print("="*50)
print("DOWNLOADING STOCK DATA")
print("="*50)

# Download SPY data (S&P 500 ETF)
print("\nDownloading SPY (S&P 500)...")
spy = yf.download('SPY', start='2019-01-01', end='2024-01-01', progress=False)

# Reset index to make Date a regular column, then save
spy_clean = spy.reset_index()

# Save to CSV with proper formatting
spy_clean.to_csv('data/spy_raw.csv', index=False)
print(f"✓ Downloaded {len(spy_clean)} days of SPY data")
print(f"✓ Saved to data/spy_raw.csv")

# Show first few rows
print("\nFirst 5 rows:")
print(spy_clean.head())

# Show column names
print("\nColumns:")
print(spy_clean.columns.tolist())

# Basic stats
print("\nBasic Statistics:")
print(f"Start date: ${spy_clean['Date'].iloc[0]}")
print(f"End date: ${spy_clean['Date'].iloc[-1]}")
print(f"Highest close: ${spy_clean['Close'].max():.2f}")
print(f"Lowest close: ${spy_clean['Close'].min():.2f}")
print(f"Average close: ${spy_clean['Close'].mean():.2f}")

print("\n" + "="*50)
print("DOWNLOAD COMPLETE!")
print("="*50)
```
