```python
import pandas as pd
import numpy as np

print("="*50)
print("PROCESSING DATA")
print("="*50)

#Method manipulations -> Calculation of all necessary parameters

data = pd.read_csv('data/spy_raw.csv') #Reading file
data['Date'] = pd.to_datetime(data['Date']) #converting string to date
data = data.set_index('Date') #setting date as index
data['returns'] = data['Close'].pct_change() # Calculate daily returns
data['MA_20'] = data['Close'].rolling(window=20).mean() #Calculate 20 last day MVA
data['volatility'] = data['returns'].rolling(window=20).std() #std dev of 20 last days

data_clean = data.dropna()# Drop rows with nothing
data_clean.to_csv('data/spy_processed.csv')

# Show summary statistics

print("\n" + "="*50)
print("SUMMARY STATISTICS")
print("="*50)

print("\nReturns:")
print(f"  Average daily return: {data_clean['returns'].mean():.4%}")
print(f"  Std dev of returns: {data_clean['returns'].std():.4%}")
print(f"  Best day: {data_clean['returns'].max():.4%}")
print(f"  Worst day: {data_clean['returns'].min():.4%}")

print("\nVolatility:")
print(f"  Average volatility: {data_clean['volatility'].mean():.4%}")
print(f"  Max volatility: {data_clean['volatility'].max():.4%}")
print(f"  Min volatility: {data_clean['volatility'].min():.4%}")

print("\nPrice:")
print(f"  Starting price: ${data_clean['Close'].iloc[0]:.2f}")
print(f"  Ending price: ${data_clean['Close'].iloc[-1]:.2f}")
print(f"  Total return: {(data_clean['Close'].iloc[-1] / data_clean['Close'].iloc[0] - 1):.2%}")

print("\n" + "="*50)
print("PROCESSING COMPLETE!")
print("="*50)

print("\nFirst 5 rows of processed data:")
print(data_clean[['Close', 'returns', 'MA_20', 'volatility']].head())
```
