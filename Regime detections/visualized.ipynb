```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data/spy_processed.csv', index_col='Date', parse_dates=True)
fig, axes = plt.subplots(3, 1, figsize=(14, 10))

# Chart 1: Price and Moving Average
axes[0].plot(data.index, data['Close'], label='SPY Price', linewidth=1.5, color='blue')
axes[0].plot(data.index, data['MA_20'], label='20-Day MA', linewidth=1.5, color='orange', linestyle='--')
axes[0].set_title('SPY Price vs 20-Day Moving Average', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Price ($)', fontsize=12)
axes[0].legend(loc='upper left')
axes[0].grid(True, alpha=0.3)

# Chart 2: Daily Returns
axes[1].bar(data.index, data['returns'], label='Daily Returns', width=1, alpha=0.6, color='green')
axes[1].axhline(y=0, color='red', linestyle='-', linewidth=0.8)
axes[1].set_title('Daily Returns', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Return (%)', fontsize=12)
axes[1].legend(loc='upper left')
axes[1].grid(True, alpha=0.3)

# Chart 3: Volatility Over Time
axes[2].plot(data.index, data['volatility'], label='20-Day Volatility', linewidth=1.5, color='red')
axes[2].fill_between(data.index, data['volatility'], alpha=0.3, color='red')
axes[2].axhline(y=data['volatility'].median(), color='black', linestyle='--', linewidth=1, label='Median Volatility')
axes[2].set_title('Rolling 20-Day Volatility', fontsize=14, fontweight='bold')
axes[2].set_ylabel('Volatility (Std Dev)', fontsize=12)
axes[2].set_xlabel('Date', fontsize=12)
axes[2].legend(loc='upper left')
axes[2].grid(True, alpha=0.3)


plt.tight_layout()
output_file = 'output/data_overview.png'
plt.savefig(output_file, dpi=300, bbox_inches='tight')
print(f"Chart saved to {output_file}")

plt.show()
```
